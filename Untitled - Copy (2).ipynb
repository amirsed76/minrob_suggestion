```python
import numpy as np 
import pandas as pd 
```

```python
class MinRoob:
    def __init__(self):
        self.probabilites = np.full((8,7) , 15/56)
        self.grids = np.full((8,7) , "*")

    def show_grid(self):
        return pd.DataFrame(self.grids)
    
    
    def show_probabilites(self):
        return pd.DataFrame(self.probabilites).round(2)
    
    
    def all_neigbors(self,x,y):
        neigbors = []
        for i in range(-1,2):
            for j in range(-1,2):
                if i==0 and j==0 :
                    pass 
                else:
                    if 0<=(x+i)<8 and 0<=(y+j)<7 :
                        if self.grids[x+i , y+j] == "*":

                            neigbors.append((x+i , y+j))

        return neigbors
    
    
    def all_0_probability_neigbors(self,x,y):
        neigbors = []
        for i in range(-1,2):
            for j in range(-1,2):
                if i==0 and j==0 :
                    pass 
                else:
                    if 0<=(x+i)<8 and 0<=(y+j)<7 :
                        if self.probabilites[x+i , y+j] == 0:

                            neigbors.append((x+i , y+j))

        return neigbors           
    def mins_in_neigbors(self,x,y):
        result = 0
        for i in range(-1,2):
            for j in range(-1,2):
                if i==0 and j==0 :
                    pass 
                else:
                    if  0<=x+i<8 and 0<=y+i<7 :

                        if self.grids[x+i , y+j] == "O":
                            result += 1 
                    
                    
        return result
    
    def counts(self):
        count_mins = 0 
        count_can_min = 0
        for i in range(0,8):
            for j in range(0,7):
                if self.grids[i,j] == "O":
                     count_mins += 1
                if self.grids[i,j] == "*":
                     count_can_min += 1
                        
        return (count_mins,count_can_min)
    
    def update_probabilites(self):
        counts = self.counts()
        probability = (15-counts[0])/counts[1]
        for x in range(0,8):
            for y in range(0,7):
                if self.grids[x,y] == "*" and self.probabilites[x,y]!=0 and self.probabilites[x,y]!=1:
                    self.probabilites[x,y] =probability
        for x in range(0,8):
            for y in range(0,7):
                try:
                    remaining_mins = int(self.grids[x,y]) - self.mins_in_neigbors(x,y)
                    active_neigbors = self.all_neigbors(x,y)
                    neibors_0 = self.all_0_probability_neigbors(x,y)
                    if (len(active_neigbors)-len(neibors_0))==0 :
                        if remaining_mins == 0:
                            probability=0
                        else:
                            probability = 1
                    else:

                        probability = (remaining_mins)/(len(active_neigbors)-len(neibors_0))
                    for cell in self.all_neigbors(x,y):
                        if probability == 0 or self.probabilites[cell[0],cell[1]]==0:
                            self.probabilites[cell[0],cell[1]] = 0
                        elif probability >= 1 or self.probabilites[cell[0],cell[1]]>=1:
                            self.probabilites[cell[0],cell[1]] = 1
                        else:
                            self.probabilites[cell[0],cell[1]] = (self.probabilites[cell[0],cell[1]] + probability)/2

                except Exception as e:
                    pass
#                     print(e)
    #                     print(f"{x} , {y} ,{self.grids[x,y]} , {e}")


    
    def determind(self , x,y,symbol):
        
        self.grids[x,y] = symbol
        self.probabilites[x,y] = None
        self.update_probabilites()
        
        
    def best_loc(self):
        max_value = -1
        best = (-1,-1)
        for i in range(8):
            for j in range(7):
                if self.probabilites[i,j] >= max_value:
                    max_value = self.probabilites[i,j]
                    best=(i,j)
#                     if  self.probabilites[i,j] >= 0.5:
#                         best = (i,j)
                        

        return best

            
            
        
        
    
```

```python
from tkinter import *
min_roob = MinRoob()

root = Tk()

text = [[None]*7 for _ in range(8)]
buttons = [[None]*7 for _ in range(8)]

def action(i, j):
    print(i,j)
    
    x = input()
    text[i][j].set(x)
    try :
        x= int(x)
    except:
        pass
    min_roob.determind(i,j,x)
    row , col = min_roob.best_loc()
    for i in range(8):
        for j in range(7):
                buttons[i][j].config(textvariable = text[i][j], width = 9, height = 5 , bg="white" )

    buttons[row][col].config(textvariable = text[row][col], width = 9, height = 5 , bg="blue" )

#     buttons[row][col] = Button(root, command = lambda i=row, j=col : action(i, j) ,bg="blue")

    print()

for i in range(8):
    for j in range(7):
        text[i][j] = StringVar()
        text[i][j].set("*")
        buttons[i][j] = Button(root, command = lambda i=i, j=j : action(i, j) )
        buttons[i][j].config(textvariable = text[i][j], width = 9, height = 5 )
        buttons[i][j].grid(row = i, column = j)
        
        
    

root.mainloop()
```
