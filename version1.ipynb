```python
import numpy as np 
import pandas as pd 
import random
```

```python
class MinRoob:
    def __init__(self):
        self.probabilites = np.full((8,7) , 15/56)
        self.grids = np.full((8,7) , "*")

    def show_grid(self):
        return pd.DataFrame(self.grids)
    
    
    def show_probabilites(self):
        return pd.DataFrame(self.probabilites).round(2)
    
    
    def all_neigbors(self,x,y):
        neigbors = []
        for i in range(-1,2):
            for j in range(-1,2):
                if i==0 and j==0 :
                    pass 
                else:
                    if 0<=(x+i)<8 and 0<=(y+j)<7 :
                        if self.grids[x+i , y+j] == "*":

                            neigbors.append((x+i , y+j))

        return neigbors
    
    
    def all_0_probability_neigbors(self,x,y):
        neigbors = []
        for i in range(-1,2):
            for j in range(-1,2):
                if i==0 and j==0 :
                    pass 
                else:
                    if 0<=(x+i)<8 and 0<=(y+j)<7 :
                        if self.probabilites[x+i , y+j] == 0:

                            neigbors.append((x+i , y+j))

        return neigbors           
    def mins_in_neigbors(self,x,y):
        result = 0
        for i in range(-1,2):
            for j in range(-1,2):
                if i==0 and j==0 :
                    pass 
                else:
                    if  0<=x+i<8 and 0<=y+j<7 :

                        if self.grids[x+i , y+j] == "X":
                            result += 1 
                    
                    
        return result
    
    
    def numbers_in_neigbors(self,x,y):
        result = 0
        for i in range(-1,2):
            for j in range(-1,2):
                if i==0 and j==0 :
                    pass 
                else:
                    if  0<=x+i<8 and 0<=y+j<7 :
                        try:
                            if int(self.grids[x+i , y+j]) in [1,2,3,4,5,6,7,8]:
                                result += 1 
                        except:
                            pass

                    
        return result
    
    def counts(self):
        count_mins = 0 
        count_can_min = 0
        for i in range(0,8):
            for j in range(0,7):
                if self.grids[i,j] == "X":
                     count_mins += 1
                if self.grids[i,j] == "*" and self.probabilites[i,j] != 0 :
                     count_can_min += 1
                        
        return (count_mins,count_can_min)
    
    def update_probabilites(self):
        counts = self.counts()
        probability = (15-counts[0])/counts[1]
        for x in range(0,8):
            for y in range(0,7):
#                 if self.grids[x,y] == "*" and self.probabilites[x,y]!=0 and self.probabilites[x,y]!=1:
                if self.grids[x,y] == "*":
                    self.probabilites[x,y] =probability
        for x in range(0,8):
            for y in range(0,7):
                try:
                    active_neigbors = self.all_neigbors(x,y)
                    neibors_0 = self.all_0_probability_neigbors(x,y)
                    remaining_mins = int(self.grids[x,y]) - self.mins_in_neigbors(x,y)
                    if (len(active_neigbors)-len(neibors_0))==0 :
                        if remaining_mins == 0:
                            probability=0
                        else:
                            probability = 1
                    else:

                        probability = (remaining_mins)/(len(active_neigbors)-len(neibors_0))

                    for cell in self.all_neigbors(x,y):
                        if probability == 0 or self.probabilites[cell[0],cell[1]]==0:
                            self.probabilites[cell[0],cell[1]] = 0
                        elif probability >= 1 or self.probabilites[cell[0],cell[1]]>=1:
                            self.probabilites[cell[0],cell[1]] = 1
                        else:
                            add = self.numbers_in_neigbors(cell[0],cell[1]) / 10
                            self.probabilites[cell[0],cell[1]] = ((self.probabilites[cell[0],cell[1]] + probability)/2)+add

                except Exception as e:
                    pass
                        
#                     print(e)
    #                     print(f"{x} , {y} ,{self.grids[x,y]} , {e}")


    
    def determind(self , x,y,symbol):
        
        self.grids[x,y] = symbol
        self.probabilites[x,y] = None
        self.update_probabilites()
        
        
    def best_loc(self):
        max_value = -1
        best = (-1,-1)
        for i in range(8):
            for j in range(7):
                if self.probabilites[i,j] > max_value:
                    max_value = self.probabilites[i,j]
                    if self.probabilites[i,j] >= 0.7:
                        best=(i,j)
                    elif random.random() <= self.probabilites[i,j]:
                        best=(i,j)
                        
        rows=list(range(8)) 
        cols = list(range(7))
        random.shuffle(rows)
        random.shuffle(cols)
        if best == (-1,-1):
            for i in rows:
                for j in cols:
                    if self.grids[i,j]=="*":
                        best = (i,j)
                        return best
               
        return best

            
            
        
        
    
```

```python
from tkinter import *
min_roob = MinRoob()

root = Tk()
text = [[None]*7 for _ in range(8)]
buttons = [[None]*7 for _ in range(8)]

def set_character(i,j):
    master = Tk()
    master.geometry("100x400")
    def bla(item):
        text[i][j].set(item)
        master.destroy()
        master.quit()
    Button(master, command = lambda item="0" : bla(item), text = '0').pack()
    Button(master, command = lambda item="1" : bla(item), text = '1').pack()
    Button(master, command = lambda item="2" : bla(item), text = '2').pack()
    Button(master, command = lambda item="3" : bla(item), text = '3').pack()
    Button(master, command = lambda item="4" : bla(item), text = '4').pack()
    Button(master, command = lambda item="5" : bla(item), text = '5').pack()
    Button(master, command = lambda item="6" : bla(item), text = '6').pack()
    Button(master, command = lambda item="7" : bla(item), text = '7').pack()
    Button(master, command = lambda item="8" : bla(item), text = '8').pack()
    Button(master, command = lambda item="X" : bla(item), text = 'X').pack()
    Button(master, command = lambda item="*" : bla(item), text = '*').pack()
   
    master.mainloop()

    
def color(value):
    if value == "X":
        return "red"
    elif value =="*":
        return "white"
    else:
        return "green"
            
def action(i, j): 
    v = set_character(i,j)
    x =text[i][j].get()
    text[i][j].set(x)
    try :
        x= int(x)
    except Exception as e:
        pass
    min_roob.determind(i,j,x)
    row , col = min_roob.best_loc()
    print(min_roob.show_grid())
    print(min_roob.show_probabilites())
    for i in range(8):
        for j in range(7):
                buttons[i][j].config(textvariable = text[i][j], width = 9, height = 5 , bg=color(text[i][j].get()) )

    buttons[row][col].config(textvariable = text[row][col], width = 9, height = 5 , bg="blue" )



for i in range(8):
    for j in range(7):
        text[i][j] = StringVar()
        text[i][j].set("*")
        buttons[i][j] = Button(root, command = lambda i=i, j=j : action(i, j) )
        buttons[i][j].config(textvariable = text[i][j], width = 9, height = 5 )
        buttons[i][j].grid(row = i, column = j)
        
        
    

root.mainloop()
```
